```python
# =========================
# 1. Importar bibliotecas
# =========================
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# 2. Carregar dataset
# =========================
df = pd.read_csv("netflix_titles[1].csv")

# Ver primeiras linhas
df.head()
```

```python
# =========================
# 3. Estrutura do dataset
# =========================
df.info()
```

```python
# =========================
# 4. Estatísticas gerais
# =========================
df.describe(include="all")
```

```python
# =========================
# 5. # Pergunta 1: Há mais filmes ou séries?
# =========================
df["type"].value_counts()
```

```python
# Gráfico: Movies vs TV Shows
df["type"].value_counts().plot(kind="bar", title="Quantidade de Movies vs TV Shows")
plt.show()
```

```python
# =========================
# 6. Pergunta 2: Quais países produzem mais conteúdo?
# (limpeza da coluna country)
# =========================

df_countries = df["country"].dropna().str.split(", ").explode()

df_countries.value_counts().head(10)
```

```python
# Gráfico: Top 10 países
df_countries.value_counts().head(10).plot(kind="bar", title="Top 10 Países com mais conteúdo")
plt.show()
```

```python
# =========================
# 7.# Pergunta 3:
# Como evolui o conteúdo por ano?

# =========================

df["release_year"].value_counts().sort_index()
```

```python
# Gráfico de evolução por ano
df["release_year"].value_counts().sort_index().plot(kind="line", title="Evolução de títulos por ano")
plt.show()
```

```python
# =========================
# 8. Pergunta 4:
# Relação entre tipo (Movie/TV Show) e ano
# =========================

df_year_type = df.groupby(["release_year", "type"]).size().unstack(fill_value=0)

df_year_type.head()
```

```python
# Gráfico: Movie vs TV Show por ano
df_year_type.plot(kind="line", title="Movies vs TV Shows por ano")
plt.show()
```
